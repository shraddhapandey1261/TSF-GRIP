## Problem Statement: Predict the percentage of an student based on the no. of study hours. This is a simple linear regression task as it involves just 2 variables. What will be predicted score if a student studies for 9.25 hrs/day?

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading data
df=pd.read_csv('http://bit.ly/w-data')
```

# Visualization

```python
#plotting distribution of scores
df.plot(x='Hours' , y='Scores' , style='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours studied')
plt.ylabel('Percentage score')
plt.show()
```

Hence, this graph shows that there is positive linear relation between number of hours studied and the percentage score.s

# Preparing the data


Dividing the dataset as X-label and y-label 

```python
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values
```

Now, as we have the attributes and labels we will split this data into training and testing sets. Here, scikit-learn's train_test_split() method is used

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Training the algorithm

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

print("Training complete.")
```

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

Hence, it clarifies that As Hours Studied increases the Percentage of Scores also increases, thus there is Positive linear Relation.

# Making Predictions

 Now that we have trained our algorithm, it's time to make some predictions.

```python
print(X_test) # Testing data - In Hours
y_pred = regressor.predict(X_test) # Predicting the scores
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df
```

# What will be the predicted score if student studied for 9.25 hrs/day?

```python
hours=9.25
test=np.array([hours])
test = test.reshape(-1,1)
own_pred = regressor.predict(test)
print("No of Hours Studied= {}".format(hours))
print("Predicted scores of students= {}".format(own_pred[0]))
```

# Evaluating the model

The final step is to evaluate the performance of algorithm. This step is particularly important to compare how well different algorithms perform on a particular dataset. For simplicity here, we have chosen the mean square error.

```python
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
```
